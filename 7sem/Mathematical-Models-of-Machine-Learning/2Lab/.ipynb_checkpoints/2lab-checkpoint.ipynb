# Модель

```python
class BaseModel(object):
    """
    Базовый класс модели.
    """
    def __init__(self):
        self.quality_metric = mean_absolute_percentage_error
        self.coefs = None

    def empirical_risk(self, coefs, X, Y):
        """
        Вычисляет эмпирический риск для заданных коэффициентов.
        Переопределяется в подклассе для конкретной модели.
        """
        raise NotImplementedError('Метод empirical_risk должен быть переопределен в подклассе.')

    def fit(self, X: pd.DataFrame, Y: pd.Series):
        """
        Метод для обучения модели, подгоняет параметры через минимизацию эмпирического риска.

        Параметры:
        X (pd.DataFrame): Матрица признаков.
        Y (pd.Series): Вектор ответов.

        Возвращает:
        self: Обученную модель.
        """
        X = X.values  # Преобразуем DataFrame в numpy массив
        Y = Y.values  # Преобразуем Series в numpy массив

        initial_coefs = np.zeros(X.shape[1] + 1)  # Начальные коэффициенты (включая свободный член)
        
        # Минимизация эмпирического риска
        res = scipy.optimize.minimize(self.empirical_risk, initial_coefs, args=(X, Y))
        
        # Сохраняем обученные коэффициенты
        self.coefs = res.x

        return self  # Это позволяет использовать цепочку вызовов, как в scikit-learn

    def predict(self, X: pd.DataFrame):
        """
        Делает предсказания на новых данных.

        Параметры:
        X (pd.DataFrame): Новые данные для предсказания.

        Возвращает:
        np.ndarray: Массив предсказаний.
        
        Исключение:
        ValueError: Если модель не обучена.
        """
        if self.coefs is None:
            raise ValueError('Модель не обучена. Сначала вызовите метод fit.')
        return np.dot(X, self.coefs[1:]) + self.coefs[0]

    def test_on_dataset(self, X_test: pd.DataFrame, Y_test: pd.Series):
        """
        Проверяет точность модели на тестовом наборе данных.

        Параметры:
        X_test (pd.DataFrame): Тестовый набор данных.
        Y_test (pd.Series): Истинные значения для тестового набора данных.

        Возвращает:
        float: Значение метрики качества на тестовом наборе данных.
        """
        assert len(X_test) == len(Y_test), 'Количество объектов в X_test и Y_test должно совпадать.'
        y_pred = self.predict(X_test)
        return self.quality_metric(Y_test, y_pred)

    def score(self, X: pd.DataFrame, y: pd.Series):
        """
        Оценивает производительность модели на предоставленных данных.

        Для задачи регрессии вычисляется R².

        Параметры:
        X (pd.DataFrame): Признаки для оценки.
        y (pd.Series): Истинные значения для оценки.

        Возвращает:
        float: Значение R² (коэффициент детерминации).
        """
        y_pred = self.predict(X)

        ss_total = np.sum((y - np.mean(y)) ** 2)  # Общая сумма квадратов отклонений
        ss_residual = np.sum((y - y_pred) ** 2)  # Сумма квадратов отклонений остатков
        r2 = 1 - (ss_residual / ss_total)
        return r2
```

```python
class M1(BaseModel):
    """
    Класс линейной модели, минимизирующей MAPE.
    """
    def __init__(self):
        super().__init__()

    def empirical_risk(self, coefs, X, Y):
        """
        Вычисляет эмпирический риск для заданных коэффициентов.
        """
        Y_pred = np.dot(X, coefs[1:]) + coefs[0]
        return self.quality_metric(Y, Y_pred)
```

```python
X = train_data.drop(columns=['y'])
Y = train_data['y']

model = M1()
model.fit(X, Y)

score = model.test_on_dataset(X, Y)
print(f'Score: {score:.4f}')
```

```python
class M1r(BaseModel):
    def __init__(self, mu=1, tau=1):
        """
        Инициализация модели M1r.
        mu — коэффициент для L1 регуляризации (штраф на абсолютные значения коэффициентов).
        tau — коэффициент для L2 регуляризации (штраф на квадраты коэффициентов).
        """
        super().__init__()
        self.mu = mu
        self.tau = tau

    def empirical_risk(self, coefs, X, Y):
        """
        Вычисляет эмпирический риск с L1 и L2 регуляризацией для заданных коэффициентов.
        """
        Y_pred = np.dot(X, coefs[1:]) + coefs[0]  # Линейное предсказание
        l1_reg = self.mu * np.sum(np.abs(coefs[1:]))  # L1 регуляризация
        l2_reg = self.tau * np.dot(coefs[1:], coefs[1:])  # L2 регуляризация
        return mean_absolute_percentage_error(Y, Y_pred) + l1_reg + l2_reg

    def get_params(self, deep=True):
        """
        Получение параметров модели для совместимости с API scikit-learn.
        """
        return {'mu': self.mu, 'tau': self.tau}

    def set_params(self, **params):
        """
        Установка параметров модели для совместимости с API scikit-learn.
        """
        for param, value in params.items():
            setattr(self, param, value)
        return self
```

```python
model = M1r()

parameters = {
    'mu': [0.0001, 0, 10000],
    'tau': [0.0001, 0, 10000]
}
grid_search = GridSearchCV(model, parameters, cv=5)
grid_search.fit(X, Y)
print('Лучшие параметры:', grid_search.best_params_)
```

По лучшим параметрам мы видим, что произошло вырождение в модель M1
