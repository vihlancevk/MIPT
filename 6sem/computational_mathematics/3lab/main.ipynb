## Метод стрельбы

- задача Коши 2-ого порядка:

$ y^{''} = x\sqrt{y} $

$ y(0) = 0 $

$ y^{'}(1) = \alpha $

- замена:

$ y_1 = y \text{, } y_2 = y^{'} $

- система:

$ y_1^{'} = y_2 $

$ y_2^{'} = x\sqrt{y_1} $

$ y_1(0) = 0 \text{, } y_2(0) = \alpha $

```python
def func_1(a):
    y1 = 0
    y2 = a
    
    x = 0
    x_end = 1
    dx = 0.005
    
    x_values = []
    y1_values = []
    y2_values = []
    
    while x <= x_end:
        k1_y1 = y2
        k1_y2 = y1**0.5
    
        k2_y1 = y2 + (dx/2) * k1_y1
        k2_y2 = (y1 + dx/2)**0.5 + (dx/2) * k1_y2
    
        k3_y1 = y2 + (dx/2) * k2_y1
        k3_y2 = (y1 + dx/2)**0.5 + (dx/2) * k2_y2
    
        k4_y1 = y2 + dx * k3_y1
        k4_y2 = (y1 + dx/2)**0.5 + (dx/2) * k3_y2
        
        y1 += (dx/6) * (k1_y1 + 2*k2_y1 + 2*k3_y1 + k4_y1)
        y2 += (dx/6) * (k1_y2 + 2*k2_y2 + 2*k3_y2 + k4_y2)
        
        x_values.append(x)
        y1_values.append(y1)
        y2_values.append(y2)
        
        x += dx
    
    return x_values, y1_values
```

```python
eps = 0.001
h = 0.001
```

```python
a = 0

x_values, y_values = func_1(a)
F = y_values[-1] - 2

while not (abs(F) < eps):
    x_values, y_values = func_1(a + h)
    F_derivative = y_values[-1] - 2
    
    a -= (F * h) / (F_derivative - F)
    
    x_values, y_values = func_1(a)
    F = y_values[-1] - 2

x_values, y_values = func_1(a)
```

```python
plt.plot(x_values, y_values, label='y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

## Метод квазилинеаризации

Метод:

$ y(x) = y_0(x) + w(x) $

$ y_0(0) = 0 \text{, } y_0(1) = 2 $

Возьмём:

$ y_0(x) = 2x $

Тогда:

$ w^{''} = \frac{xw}{2\sqrt{2x + w}} + x\sqrt{2x + w} $

$ w(0) = w(1) = 0 $

- задача Коши 2-ого порядка:

$ w^{''} = \frac{xw}{2\sqrt{2x + w}} + x\sqrt{2x + w} $

$ w(0) = 0 $

$ w^{'}(1) = \alpha $

- замена:

$ w_1 = w \text{, } w_2 = w^{'} $

- система:

$ w_1^{'} = w_2 $

$ w_2^{'} = \frac{xw_1}{2\sqrt{2x + w_1}} + x\sqrt{2x + w_1} $

$ w_1(0) = 0 \text{, } w_2(0) = \alpha $

```python
def func_2(a):
    w1 = 1e-6
    w2 = a
    
    x = 0
    x_end = 1
    dx = 0.005
    
    x_values = []
    w1_values = []
    w2_values = []
    
    while x <= x_end:
        k1_w1 = w2
        k1_w2 = (x * w1) / (2 * ((2*x + w1)**0.5)) + x * ((2*x + w1)**0.5)
    
        k2_w1 = y2 + (dx/2) * k1_w1
        k2_w2 = ((x + (dx/2)) * (w1 + (dt/2)*k1_w2)) / (2 * ((2*(x + (dx/2)) + (w1 + (dt/2)*k1_w2))**0.5)) + \
        (x + (dx/2)) * ((2*(x + (dx/2)) + (w1 + (dt/2)*k1_w2))**0.5)
    
        k3_w1 = y2 + (dx/2) * k2_w1
        k3_w2 = ((x + (dx/2)) * (w1 + (dt/2)*k2_w2)) / (2 * ((2*(x + (dx/2)) + (w1 + (dt/2)*k2_w2))**0.5)) + \
        (x + (dx/2)) * ((2*(x + (dx/2)) + (w1 + (dt/2)*k2_w2))**0.5)
    
        k4_w1 = y2 + dx * k3_w1
        k4_w2 = ((x + (dx/2)) * (w1 + (dt/2)*k3_w2)) / (2 * ((2*(x + (dx/2)) + (w1 + (dt/2)*k3_w2))**0.5)) + \
        (x + (dx/2)) * ((2*(x + (dx/2)) + (w1 + (dt/2)*k3_w2))**0.5)
        
        w1 += (dx/6) * (k1_w1 + 2*k2_w1 + 2*k3_w1 + k4_w1)
        w2 += (dx/6) * (k1_w2 + 2*k2_w2 + 2*k3_w2 + k4_w2)
        
        x_values.append(x)
        w1_values.append(w1)
        w2_values.append(w2)
        
        x += dx
    
    return x_values, w1_values
```

```python
eps = 0.001
h = 0.001
```

```python
a = 0

x_values, w_values = func_2(a)
F = w_values[-1]

while not (abs(F) < eps):
    x_values, w_values = func_2(a + h)
    F_derivative = w_values[-1]
    
    a -= (F * h) / (F_derivative - F)
    
    x_values, w_values = func_2(a)
    F = w_values[-1]

x_values, w_values = func_2(a)
```

```python
plt.plot(x_values, w_values, label='w(x)')
plt.xlabel('x')
plt.ylabel('w')
plt.legend()
```

```python
y0_values = 2*np.array(x_values)
y_values = y0_values + w_values
```

```python
plt.figure(2)
plt.plot(x_values, y_values, label='y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
